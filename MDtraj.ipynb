```python
import mdtraj as md
import numpy as np

t = md.load('coord_4.nc', top='top.prmtop')
top = t.topology

#pair = np.array([[1,2]])
#md.compute_distances(t, pair)
```

```python
nit_idx = []
ox_idx = []
for residues in top.residues:
    if residues.n_atoms > 20:
        for atom in residues.atoms:
            if atom.element.symbol == 'O':
                ox_idx.append(atom.index)
            if atom.element.symbol == 'N':
                nit_idx.append(atom.index)
```

```python
nit_ox_idx = sorted(nit_idx + ox_idx)
nit_ox = np.reshape(nit_ox_idx,(int(len(nit_ox_idx)/4),4))

#nit = np.reshape(nit_idx,(int(len(nit_idx)/2),2))

for f in enumerate(nit_ox):
    query_indices=np.array(f[1])
    x = np.delete(nit_ox, (f[0]), axis=0)
    haystack_indices = x.flatten()
    neigh = md.compute_neighbors(t[500], 1, query_indices, haystack_indices, periodic=True)
    res_neigh = []
    for f in neigh[0]:
        res = top.atom(f).residue
        if res not in res_neigh:
            res_neigh.append(res)
            
    print(res_neigh)
        
    
```

```python
hbonds = md.wernet_nilsson(t, exclude_water=True, periodic=True, sidechain_only=False)
```

```python
label = lambda hbond : '%s -- %s' % (t.topology.atom(hbond[0]), t.topology.atom(hbond[2]))
count = 1
for i in range(len(hbonds)):
    if len(hbonds[i]) > 1:
        for f in range(len(hbonds[i])):
            print('frame',count, label(hbonds[i][f]))
    count += 1
```

```python
for frame in range(t.n_frames):
    for res in enumerate(nit_ox):
        other_res = np.delete(nit_ox, (res[0]), axis=0)
        other_res = other_res.flatten()
        for atom1 in res:
            for atom2 in other_res:
                dist1_2 = md.compute_distances(t[frame], [[atom1,atom2]], periodic=True, opt=True)
                if dist1_2 < 3:
                    print(top.atom(atom1).residue, top.atom(atom2).residue)
        

    #md.compute_distances(t, [[200,300]], periodic=True, opt=True)
```
